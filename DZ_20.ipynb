Задание №1 Class “Книга”
Создайте класс Книга с атрибутами:
название (строка),
автор (строка),
год_издания (целое число),
количество_страниц (целое число),
цена (вещественное число),
Добавьте в класс Книга метод рассчитать_стоимость,
который будет возвращать стоимость книги, исходя из
ее цены и количества страниц. Формула для расчета
стоимости: цена * количество_страниц / 100.
Добавьте в класс Книга метод
изменить_количество_страниц, который будет
принимать новое количество страниц и изменять
соответствующий атрибут.

```python
class Книга:
    def __init__(self, название, автор, год_издания, количество_страниц, цена):
        self.название = название
        self.автор = автор
        self.год_издания = год_издания
        self.количество_страниц = количество_страниц
        self.цена = цена

    def рассчитать_стоимость(self):
        """Метод для расчета стоимости книги."""
        return self.цена * self.количество_страниц / 100

    def изменить_количество_страниц(self, новое_количество_страниц):
        """Метод для изменения количества страниц книги."""
        self.количество_страниц = новое_количество_страниц

    def __str__(self):
        """Метод для удобного отображения информации о книге."""
        return (f'Книга: {self.название}, Автор: {self.автор}, '
                f'Год издания: {self.год_издания}, '
                f'Количество страниц: {self.количество_страниц}, '
                f'Цена: {self.цена}.')


if __name__ == "__main__":
    книга1 = Книга("Война и мир", "Лев Толстой", 1869, 1225, 300.50)
    
    print(книга1)
    стоимость = книга1.рассчитать_стоимость()
    print(f'Стоимость книги: {стоимость:.2f} руб.')

    книга1.изменить_количество_страниц(1300)
    print(f'Новое количество страниц: {книга1.количество_страниц}')
    
    новая_стоимость = книга1.рассчитать_стоимость()
    print(f'Новая стоимость книги: {новая_стоимость:.2f} руб.')
```

Задание №2 Class “Counter”
Создайте класс Счетчик, который будет моделировать
счетчик, используемый для подсчета чего-либо. Класс
должен иметь следующие свойства и методы:
Приватное свойство _значение, которое хранит
текущее значение счетчика.
Метод увеличить(), который увеличивает значение
счетчика на 1.
Метод уменьшить(), который уменьшает значение
счетчика на 1.
Метод получить_значение(), который возвращает
текущее значение счетчика.
Метод сбросить(), который сбрасывает значение
счетчика на 0.
Дополнительные требования:
Значение счетчика не может быть меньше 0.
При попытке уменьшить значение счетчика ниже 0,
оно должно оставаться равным 0.

```python
class Счетчик:
    def __init__(self):
        self._значение = 0

    def увеличить(self):
        """Метод для увеличения значения счетчика на 1."""
        self._значение += 1

    def уменьшить(self):
        """Метод для уменьшения значения счетчика на 1, не позволяя ему стать меньше 0."""
        if self._значение > 0:
            self._значение -= 1

    def получить_значение(self):
        """Метод для получения текущего значения счетчика."""
        return self._значение

    def сбросить(self):
        """Метод для сброса значения счетчика на 0."""
        self._значение = 0


if __name__ == "__main__":
    счетчик = Счетчик()
    
    счетчик.увеличить()
    счетчик.увеличить()
    print(f'Текущее значение счетчика: {счетчик.получить_значение()}')

    счетчик.уменьшить()
    print(f'Текущее значение счетчика: {счетчик.получить_значение()}')

    счетчик.уменьшить()
    счетчик.уменьшить()
    print(f'Текущее значение счетчика: {счетчик.получить_значение()}')

    счетчик.сбросить()
    print(f'Текущее значение счетчика: {счетчик.получить_значение()}')
```

Задание №3 Class “BankAcc”
Создайте класс БанковскийСчет, который будет
моделировать банковский счет. Класс должен иметь
следующие свойства и методы:
Приватное свойство _баланс, которое хранит текущий
баланс счета.
Метод положить_деньги(сумма), который увеличивает
баланс на указанную сумму.
Метод снять_деньги(сумма), который уменьшает
баланс на указанную сумму. Если на счете
недостаточно средств, должно выводиться сообщение
об ошибке.
Метод получить_баланс(), который возвращает
текущий баланс счета.
Метод перевести_деньги(другой_счет, сумма),
который переводит указанную сумму с текущего счета
на другой счет. Если на текущем счете недостаточно
средств, должно выводиться сообщение об ошибке.
Дополнительные требования:
Баланс не может быть отрицательным.
Сумма, передаваемая в методы, должна быть
положительной.

```python
class БанковскийСчет:
    def __init__(self):
        self._баланс = 0

    def положить_деньги(self, сумма):
        if сумма > 0:
            self._баланс += сумма
            print(f'Добавлено {сумма} на счет. Новый баланс: {self._баланс}')
        else:
            print('Ошибка: сумма должна быть положительной.')

    def снять_деньги(self, сумма):
        if сумма > 0:
            if сумма <= self._баланс:
                self._баланс -= сумма
                print(f'Снято {сумма} с счета. Новый баланс: {self._баланс}')
            else:
                print('Ошибка: недостаточно средств на счете.')
        else:
            print('Ошибка: сумма должна быть положительной.')

    def получить_баланс(self):
        return self._баланс

    def перевести_деньги(self, другой_счет, сумма):
        if сумма > 0:
            if сумма <= self._баланс:
                self._баланс -= сумма
                другой_счет.положить_деньги(сумма)
                print(f'Переведено {сумма} на другой счет. Новый баланс: {self._баланс}')
            else:
                print('Ошибка: недостаточно средств на счете.')
        else:
            print('Ошибка: сумма должна быть положительной.')


if __name__ == "__main__":
    счет1 = БанковскийСчет()
    счет2 = БанковскийСчет()

    счет1.положить_деньги(1000)
    счет1.снять_деньги(500)
    print(f'Баланс счета 1: {счет1.получить_баланс()}')

    счет1.перевести_деньги(счет2, 300)
    print(f'Баланс счета 1: {счет1.получить_баланс()}')
    print(f'Баланс счета 2: {счет2.получить_баланс()}')

    счет1.снять_деньги(1000)
    счет1.положить_деньги(-200)
```
